```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")
```

```python
songs = pd.read_csv('ma_songs_lyrics.csv', index_col=0)
```

```python
songs = songs[  (songs['lyrics'] != '(lyrics not available)')
              & (songs['lyrics'] != '(Instrumental)')
              & ~songs['lyrics'].str.contains('[^\x00-\x7f]')]
songs = songs.reset_index(drop='True')
bands = pd.read_csv('ma_bands_data.csv', index_col=0)
songs = pd.merge(bands, songs, left_on='name', right_on='band_name')
#songs.head(10)
songs.shape[0]
```

```python
blacksongs=songs[songs['genre'].str.contains('Black')]
display(blacksongs)
```

```python
genres=[]
for i in range(blacksongs.shape[0]):
    genres.append(blacksongs.iloc[i]['genre'])
len(genres)
```

```python
genres=[]
for i in range(blacksongs.shape[0]):
    genres.append(blacksongs.iloc[i]['genre'])
#display(genres)

for idx in range(len(genres)):
    genres[idx]=genres[idx].replace('(now);','/')
    genres[idx]=genres[idx].replace('(now)','')
    genres[idx]=genres[idx].replace('(early);','/')
    genres[idx]=genres[idx].replace(' early);','/')
    genres[idx]=genres[idx].replace('(early/later);','/')
    genres[idx]=genres[idx].replace('(mid);','/')
    genres[idx]=genres[idx].replace('(mid)','')
    genres[idx]=genres[idx].replace('(later)','')
    genres[idx]=genres[idx].replace(';','/')
    genres[idx]=genres[idx].replace(',','/')
    

count={}
for g in genres:
    count[g]=count.get(g,0)+1
g1=count.keys()
display(g1)

lis1=[]
for gg in g1:
    gg=gg.split('/')
    #for j in range(len(gg)):
        #if gg[j].find('Metal')==-1:
            #gg[j]=gg[j]+' Metal'
    
    for k in gg:
        #print(k)
        if k not in lis1:
            lis1.append(k)
lis1
```

```python
lis2=lis1.copy()

print('\n\n')
#print(lis2)    

for n in range(len(lis2)):
    lis2[n]=lis2[n].strip()
    if lis2[n].rfind('Metal')==len(lis2[n])-len('Metal'):
        lis2[n]=lis2[n].replace(' Metal','')
lis2
```

```python
black_genre=[]
for p in lis2:
    if p.find('Black')!=-1:
        black_genre.append(p)
print(len(black_genre))
black_genre
```

```python
song_num={}

for gen in black_genre:
    this_genre=blacksongs[blacksongs['genre'].str.contains(gen)]
    song_num[gen]=this_genre.shape[0]
    
song_num
```

```python
len(song_num)
```

```python
df=pd.DataFrame.from_dict(song_num,orient='index')
df.columns=['numbers']
df
```

```python
df0=df.iloc[16:]
df0['numbers'].sum()
```

```python
df=df.sort_values(by=['numbers'],ascending=False)
df.to_csv('Black Metal Types.csv')
```

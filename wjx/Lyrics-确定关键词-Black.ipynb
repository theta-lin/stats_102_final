```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import warnings 
warnings.filterwarnings("ignore")

songs = pd.read_csv('ma_songs_lyrics.csv', index_col=0)
```

```python
songs = songs[  (songs['lyrics'] != '(lyrics not available)')
              & (songs['lyrics'] != '(Instrumental)')
              & ~songs['lyrics'].str.contains('[^\x00-\x7f]')]
songs = songs.reset_index(drop='True')
bands = pd.read_csv('ma_bands_data.csv', index_col=0)
songs = pd.merge(bands, songs, left_on='name', right_on='band_name')
songs.head(10)
```

```python
genres=[]
for i in range(songs.shape[0]):
    genres.append(songs.iloc[i]['genre'])
#display(genres)

count={}
for g in genres:
    count[g]=count.get(g,0)+1
g1=count.keys()
#display(g1)

lis1=[]
for gg in g1:
    gg=gg.split('/')
    for j in range(len(gg)):
        if gg[j].find('Metal')==-1:
            gg[j]=gg[j]+' Metal'
    
    for k in gg:
        #print(k)
        if k not in lis1:
            lis1.append(k)
lis1
```

```python
lis2=lis1.copy()
#print(lis1)
for k in range(len(lis1)):
    #print(lis1[k])
    if lis1[k].find('early')!=-1 or lis1[k].find('mid')!=-1 or lis1[k].find('later')!=-1:
        lis2.remove(lis1[k])

print('\n\n')
#print(lis2)    

for n in range(len(lis2)):
    lis2[n]=lis2[n].replace(' Metal','')
lis2
```

```python
black=[]
for gen in lis2:
    if gen.find('Black')!=-1 and gen.find(',')==-1 and gen.find(';')==-1 and gen.find('with')==-1:
        black.append(gen)

black.remove('Technical Progressive Black')
black.remove('Melodic Pagan Black')
black.remove('Epic Pagan Black')
black.remove('Epic Folk Black')
black.remove('Raw Pagan Black')
black.remove('Experimental Psychedelic Black')



for i in range(len(black)):
    if black[i].find('Blackened')!=-1 or black[i].find('Post-Black')!=-1:
        black[i]='Blackened'
    if black[i].find('Symphonic')!=-1:
        black[i]='Symphonic'
    if black[i].find('Atmospheric')!=-1:
        black[i]='Atmospheric'
    if black[i].find('Ambient')!=-1:
        black[i]='Ambient'
    if black[i].find('Avant-garde')!=-1:
        black[i]='Avant-garde'
    if black[i].find('Depressive')!=-1:
        black[i]='Depressive'
    if black[i]=='Black Folk':
        black[i]='Folk Black'
    if black[i].find('roll')!=-1:
        black[i]=black[i].replace('roll','Roll')
        
    if black[i]!='Black \'n\' Roll' and black[i]!='Black' and black[i]!='Blackened':
        black[i]=black[i].replace(' Black','')
    if black[i]=='Black':
        black[i]='Black Metal'
        
black1=[]
for gen in black:
    if gen not in black1:
        black1.append(gen)


print(len(black1))
black1
```

```python
Black=songs[songs['genre'].str.contains('Black')]
```

```python
trivial=['i','me','my','mine','you','your','yours','he','his','him','she','her','hers','they'
        'them','their','theirs','we','us','our','ours','it','its',
        'and','or','but','in','on','at','to','into','from','since','for','by','as','this','that'
         'those','these','the','yes','no','so','if','a','of','with',
        'is','am','are','be','was','were','i\'m']

lis3=[]
for a in range(len(black1)):
    lis3.append(0)
    
for b in range(len(black1)):
    #print(lis2[b])
    df=Black[Black['genre'].str.contains(black1[b])]
    words = df['lyrics'].str.lower().str.findall("[a-z][a-z'-]*").explode()
    word_counts = words.value_counts()
    word_counts=word_counts.nlargest(20)
    
    nontriv = list(filter(lambda x: x not in trivial, words))
    nontriv=nontriv[:10]
    lis3[b]=nontriv
lis3
```

```python
KWdf=pd.DataFrame()
for i in range(len(black1)):
    KWdf[black1[i]]=lis3[i]
KWdf=KWdf.T
display(KWdf)

KWdf.to_csv("Keywords-BlackMetal.csv")
```
